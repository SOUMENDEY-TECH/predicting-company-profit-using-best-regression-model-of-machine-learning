#### As tasks given that the R&D Spend, Administration Cost and Marketing Spend of 50 Companies are given along with the profit earned. So, x values are R&D Spend, Administration Cost and Marketing Spend and y values are Profit

```python
x = companies_data.iloc[:,:-1].values
y = companies_data.iloc[:,3].values
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
x[:,2] = labelencoder.fit_transform(x[:,2])
x1 = pd.DataFrame(x)
x1
```

## i) Construct Different Regression algorithms
#### I am used i) Linear Regression, ii) Random Forest Regresson, iii) Decision Tree Regresson, iv) Lasso algorithms

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso
```

## ii) Divide the data into train set and test set
#### Split the data into train data and test data with the ratio of 70% data as train data and remaining 30% data as test data.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.7,random_state=0)
x_train = pd.DataFrame(x_train, columns=['R&D Spend', 'Administration', 'Marketing Spend'])
x_train[0:10]
```

```python
x_test = pd.DataFrame(x_test, columns=['R&D Spend', 'Administration', 'Marketing Spend'])
x_test[0:10]
```

```python
y_train[0:10]
```

```python
y_test[0:10]
```

## iii) Calculate different regression metrics

#### Calculate the Linear Regression metrix
#### And calculate the score of test and train data of the Linear Regression model

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
testing_data_linerregression_score = lr.score(x_test,y_test)
print('testing data model score is', testing_data_linerregression_score)
training_data_linerregression_score = lr.score(x_train,y_train)
print('training data model score', training_data_linerregression_score)
```

#### Compare the Actual Values with Estimates Values of the Linear Regression model

```python
data_comp = pd.DataFrame({'Actual Values': y_test.flatten(), 'Estimates': y_pred.flatten()})
data_comp
```

#### Check the R2 Score, Mean Squared Error, Root Mean Squared Error, Mean Absolute error of the Linear Regression model

```python
linear_reg_r2 = r2_score(y_test, y_pred)
print('R2 Score of the linear regression model is :', linear_reg_r2)
linear_reg_mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error of the linear regression model is :', linear_reg_mse)
linear_reg_rmse = math.sqrt(linear_reg_mse)
print('Root Mean Squared Error of the linear regression model is :', linear_reg_rmse)
linear_reg_mae = mean_absolute_error(y_test, y_pred)
print('Mean Absolute error of the linear regression model is :', linear_reg_mae)
```

#### Calculate the Random Forest Regresson metrix
#### And calculate the score of test and train data of the Random Forest Regresson model

```python
rfr = RandomForestRegressor()
rfr.fit(x_train,y_train)
y_pred = rfr.predict(x_test)
testing_data_randomforestregresson_score = rfr.score(x_test,y_test)
print('testing data model score is', testing_data_randomforestregresson_score)
training_data_randomforestregresson_score = rfr.score(x_train,y_train)
print('training data model score', training_data_randomforestregresson_score)
```

#### Compare the Actual Values with Estimates Values of the Random Forest Regresson model

```python
data_comp = pd.DataFrame({'Actual Values': y_test.flatten(), 'Estimates': y_pred.flatten()})
data_comp
```

#### Check the R2 Score, Mean Squared Error, Root Mean Squared Error, Mean Absolute error of the Random Forest Regresson model

```python
rand_forest_reg_r2 = r2_score(y_test, y_pred)
print('R2 Score of the Random Forest Regresson model is :', rand_forest_reg_r2)
rand_forest_reg_mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error of the Random Forest Regresson model is :', rand_forest_reg_mse)
rand_forest_reg_rmse = math.sqrt(rand_forest_reg_mse)
print('Root Mean Squared Error of the Random Forest Regresson model is :', rand_forest_reg_rmse)
rand_forest_reg_mae = mean_absolute_error(y_test, y_pred)
print('Mean Absolute error of the Random Forest Regresson model is :', rand_forest_reg_mae)
```

#### Calculate the Decision Tree Regresson matrix
#### And calculate the score of test and train data of the Decision Tree Regresson model

```python
dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
y_pred = dtr.predict(x_test)
testing_data_decisiontreeregresson_score = dtr.score(x_test,y_test)
print('testing data model score is', testing_data_decisiontreeregresson_score)
training_data_decisiontreeregresson_score = dtr.score(x_train,y_train)
print('training data model score', training_data_decisiontreeregresson_score)
```

#### Compare the Actual Values with Estimates Values of the Decision Tree Regresson model

```python
data_comp = pd.DataFrame({'Actual Values': y_test.flatten(), 'Estimates': y_pred.flatten()})
data_comp
```

#### Check the R2 Score, Mean Squared Error, Root Mean Squared Error, Mean Absolute error of the Decision Tree Regresson model

```python
dec_tree_reg_r2 = r2_score(y_test, y_pred)
print('R2 Score of the Decision Tree Regresson model is :', dec_tree_reg_r2)
dec_tree_reg_mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error of the Decision Tree Regresson model is :', dec_tree_reg_mse)
dec_tree_reg_rmse = math.sqrt(dec_tree_reg_mse)
print('Root Mean Squared Error of the Decision Tree Regresson model is :', dec_tree_reg_rmse)
dec_tree_reg_mae = mean_absolute_error(y_test, y_pred)
print('Mean Absolute error of the Decision Tree Regresson model is :', dec_tree_reg_mae)
```

#### Calculate the Lasso Regression matrix
#### And calculate the score of test and train data of the Lasso Regression model

```python
las = Lasso()
las.fit(x_train,y_train)
y_pred = las.predict(x_test)
testing_data_lasso_score = las.score(x_test,y_test)
print('testing data model score is', testing_data_lasso_score)
training_data_lasso_score = las.score(x_train,y_train)
print('training data model score', training_data_lasso_score)
```

#### Compare the Actual Values with Estimates Values of the Lasso Regression model

```python
data_comp = pd.DataFrame({'Actual Values': y_test.flatten(), 'Estimates': y_pred.flatten()})
data_comp
```

#### Check the R2 Score, Mean Squared Error, Root Mean Squared Error, Mean Absolute error of the Lasso Regression model

```python
lasso_r2 = r2_score(y_test, y_pred)
print('R2 Score of the Lasso Regression model is :', lasso_r2)
lasso_mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error of the Lasso Regression model is :', lasso_mse)
lasso_rmse = math.sqrt(lasso_mse)
print('Root Mean Squared Error of the Lasso Regression model is :', lasso_rmse)
lasso_mae = mean_absolute_error(y_test, y_pred)
print('Mean Absolute error of the Lasso Regression model is :', lasso_mae)
```

## iv) Choose the best model

```python
res = pd.DataFrame({'Model':['Linear Regression Model', 'Random Forest Regresson Model', 'Decision Tree Regresson Model', 'Lasso Regression Model' ],
                   'R2': [linear_reg_r2, rand_forest_reg_r2, dec_tree_reg_r2, lasso_r2],
                   'MSE': [linear_reg_mse, rand_forest_reg_mse, dec_tree_reg_mse, lasso_mse],
                   'MAE': [linear_reg_mae, rand_forest_reg_mae, dec_tree_reg_mae, lasso_mae]})
res
```

#### For choosing best model R2 Score should be maximum and Mean Squared Error rank, Mean Absolute Error rank should be minimum

```python
res['Mean Squared Error rank'] = res['MSE'].rank(ascending=True, method='min')
res['Mean Absolute Error rank'] = res['MAE'].rank(ascending=True, method='min')
res['R2 Score rank'] = res['R2'].rank(ascending=True, method='min')
res['Total rank'] = res['Mean Squared Error rank'] + res['Mean Absolute Error rank'] + res['R2 Score rank']
res_sortby_rank = res.sort_values('Total rank')
best_model_rank = res_sortby_rank.iloc[0]['Model']
print('best model :', best_model_rank)
```
